```python
from tabulate import tabulate
from IPython.display import display, Markdown
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

headers = ['block size', 'sequential time, s', 'parallel first loop time, s', 'parallel second loop time, s']
# display(Markdown('## results'))


file = open('out.txt', 'r')
for line in file:
    line = line.split()
    if(len(line) < 2) :
        break
    size = line[0]
    rows = line[1]
    title = 'matrix {}x{}'.format(size, size)
    display(Markdown('### {}'.format(title)))
    table = [file.readline().split() for _ in range(int(rows))]
    #display(Markdown(tabulate(table, headers, tablefmt='pipe')))
    
    data = np.array(table)
    labels = ['block size','1  thread', '2 threads', '4 threads']
    print(tabulate(data,labels))
    block_sizes = data[:,0]
    #ax = plt.subplot(111)
    #plt.xlabel('Block size')
    #plt.ylabel('Elapsed time, seconds')
    plt.figure(figsize=(10,5))
    #ax.set_xticks(range(2,2000,10))
    plt.plot(block_sizes,  list(map(float, data[:,1])), label=labels[1], color = 'blue')
    plt.plot(block_sizes,  list(map(float, data[:,2])), label=labels[2], color = 'green')
    plt.plot(block_sizes,  list(map(float, data[:,3])), label=labels[3], color = 'red')
    plt.show()
```
